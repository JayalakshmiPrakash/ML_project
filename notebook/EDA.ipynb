Insights
Group E students have scored the highest marks.

Group A students have scored the lowest marks.

Students from a lower Socioeconomic status have a lower avg in all course subjects

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(5, 5))

# Plot the count of different types of lunch
plt.hist(df['lunch'], bins='auto', color='purple')

# Set the title and axis labels
plt.title('Comparison of different types of lunch', fontweight=10, fontsize=10)
plt.xlabel('Types of lunch')
plt.ylabel('Count')

# Display the plot
plt.show()
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Plot the count of students based on test preparation course, grouped by parental level of education
sns.countplot(x='parental_level_of_education', data=df, hue='test_preparation_course', palette='bright', saturation=0.95, ax=axes[0])

# Set the title for the first subplot
axes[0].set_title('Students vs Test Preparation Course', color='black', size=25)

# Rotate the x-axis tick labels by 45 degrees
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)

# Plot the count of students based on lunch type, grouped by parental level of education
sns.countplot(x='parental_level_of_education', data=df, hue='lunch', palette='bright', saturation=0.95, ax=axes[1])

# Set the title for the second subplot
axes[1].set_title('Students vs Lunch Type', color='black', size=25)

# Rotate the x-axis tick labels by 45 degrees
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)

# Add labels to the bars in both subplots
for ax in axes:
    for container in ax.containers:
        ax.bar_label(container, color='black', size=20)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Insights

Students who get Standard Lunch tend to perform better than students who got free/reduced lunch
